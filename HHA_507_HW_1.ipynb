##### JSON file #1

##### **Description of the JSON file:** This data represents the numbers, percentages, and rates of referrals to child protectives services by state.

- Screened-in referrals is defined as an allegation of child maltreatment that met the state’s standards for acceptance and became a report
- Screened-out referrals is defined as an allegation of child maltreatment that did not meet the state’s standards for acceptance as a report

| Column | Description |
|---|---|
| state | Name of the state |
| screened_in_referrals_reports | Number of screened-in referrals |
| screened_out_referrals | Number of screened-out referrals |
| total_referrals | Total number of screened-in and screened-out referrals |
| screened_in_referrals_reports_1 | Percentage of screened-in referrals |
| screened_out_referrals_percent | Percentage of screened-out referrals |
| total_referrals_rate_per | The frequency by each state of referrals alleging child maltreatment per 100,000 children |
